```python
#Quitar las casillas sin valores (nulos)

df['quantity'] = df['quantity'].fillna(0)

df['order_id'] = pd.to_numeric(df['order_id'], errors='coerce')
ultimo_id = df['order_id'].max()


for i in df.index:
    if pd.isna(df.at[i, 'order_id']):
        ultimo_id += 1
        df.at[i, 'order_id'] = ultimo_id

df['order_id'] = df['order_id'].astype(int)

df.head(20)
```

```python
# Cambiar de tipos de datos

# orden de fecha y el cambio del signo
df['order_date'] = df['order_date'].astype(str).str.replace('/', '-', regex=False)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=False, errors='coerce')

#Arreglar errores de signos ( puntos y comas )
df['unit_price'] = df['unit_price'].astype(str).str.replace('$', '', regex=False)
df['unit_price'] = df['unit_price'].str.replace(',', '.', regex=False)
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')

df['unit_price'] = df['unit_price'].fillna(df.groupby('product_id')['unit_price'].transform('median'))
df['unit_price'] = df['unit_price'].fillna(0)


df['customer_name'] = df['customer_name'].astype(str).str.title().str.strip()



# Arreglar codigo del pais (seguir un unico formato de dos letras)
df['country'] = df['country'].astype(str).str.strip()

mapa_paises = {
    'Spain': 'ES', 'spain': 'ES', 'España': 'ES', 'ESP': 'ES', 
    'France': 'FR', 'france': 'FR', 'Francía': 'FR', 
    'Germany': 'DE', 'Deutschland': 'DE', 'Alemania': 'DE', 'germany': 'DE',
    'Italy': 'IT', 'Italia': 'IT', 'italy': 'IT',
    'UK': 'GB', 'United Kingdom': 'GB', 'Reino Unido': 'GB',
    'Portugal': 'PT'
}
df['country'] = df['country'].replace(mapa_paises)
df['country'] = df['country'].str.upper()

#Arreglar el precio total despues de haber arreglado cantidad nula (substituirla por 0)
df['total_price'] = df['quantity'] * df['unit_price']

df.head(5)
```

```python
#SQL
```

```python
#Creación de las tablas

#Creamos las tabla paa clientes
dim_clientes = df[["customer_id", "customer_name"]].drop_duplicates().reset_index(drop=True)

#Creamos las tabla para Productos
dim_productos = df[["product_id", "product_name", "category"]].drop_duplicates().reset_index(drop=True)

#Creamos la tabla con los datos de los pedidos
fact_pedidos = df[["order_id", "customer_id", "product_id", "order_date", "quantity", "unit_price", "total_price", "payment_method",
                  "country", "city", "status", "update_time"]].copy()
```

```python
#Conectamos las tablas con la base de datos
conn = sqlite3.connect("warehouse_pandas.db")
```

```python
#Añadimos las tablas creadas anteriormente a la base de datos y las reemplazamos si ya existen
dim_clientes.to_sql("dim_clientes", conn, if_exists="replace", index=False)
dim_productos.to_sql("dim_productos", conn, if_exists="replace", index=False)
fact_pedidos.to_sql("fact_pedidos", conn, if_exists="replace", index=False)
```

```python
#Prueba de consulta
df_fact = pd.read_sql("SELECT * FROM fact_pedidos", conn)
df_fact.head()  
```

```python
#Cerramos la conexión con la base de datos.
conn.close()
```
